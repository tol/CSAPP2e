```
#@title Code can be collapsed and hidden (double click to see it) { display-mode: "form" }

import matplotlib.pyplot as plt
import numpy as np

f, ax = plt.subplots(1, 2)
ax[0].plot(np.sin(np.arange(100) * 0.1) + np.cos(np.arange(100) * 0.3))

img = np.expand_dims(np.arange(100) ** 2, 0) + np.expand_dims(np.arange(100) ** 2, 1)
img = np.sqrt(img)
img = np.sin(img * 0.2)
ax[1].imshow(img)
ax[1].grid(False)

None
```
